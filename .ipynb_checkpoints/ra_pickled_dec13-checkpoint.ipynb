```python
import pickle
import random
import sys, os
# from collections import defaultdict
from datetime import datetime
from pytz import timezone
import pytz
```

```python
infile = open("./dec13.pkl","rb")
data = pickle.load(infile)
```

```python
time_format = "%Y-%m-%dT%H:%M:%S"
xrange = range(-10,10)
yrange = range(-5,5)
```

```python
", ".join(set(map(lambda x: x["eventKey"],data)))
```

```python
def convert_to_datetime(timeUTCstring):
    return datetime.strptime(timeUTCstring,time_format)

def convert_to_pacific(timeUTCobj):
    timePT = pytz.utc.localize(timeUTCobj).astimezone(pytz.timezone('US/Pacific'))
    return timePT.strftime(time_format)

elt = random.choice(data)
t_raw_string = elt["timeUTC"]
t_datetime   = convert_to_datetime(elt["timeUTC"])
t_datetimePT = convert_to_pacific(t_datetime)
print("STR UTC:", t_raw_string)
print("DT  UTC:", t_datetime) 
print("STR  PT:", t_datetimePT)
```

```python
plants = []
plants_at_time = []

current_grid = init_grid()
last_grid = init_grid()

# print_grid(current_grid)

plant_keys = ["playerWhoPlanted","x","y","plantType","id"]
for elt in data:
    if ("plants" in elt) and ("null" != elt["plants"]):
        current_plants = []
        datetimeUTC = convert_to_datetime(elt["timeUTC"])
        timePT = convert_to_pacific(datetimeUTC)
        for plant in elt["plants"]:
            cplant = elt["plants"][plant]
            a_plant = {k: cplant[k] for k in plant_keys}
            a_plant["datetimeUTC"] = datetimeUTC
            a_plant["timePT"] = timePT
            current_plants.append(a_plant)
        if current_plants:
            plants.extend(current_plants)
            last_grid = current_grid
            current_grid = init_grid()
            for plant in current_plants:
                current_grid[plant["x"]][plant["y"]] = plant
            plants_at_time.append([timePT,current_grid])
```

```python
from collections import defaultdict
init_grid = lambda: defaultdict(lambda: defaultdict(dict))
#lambda : {x: {y: "." for y in yrange} for x in xrange}

def grid2string(p_grid):
    out = ""
    for y in yrange:
        for x in xrange:
            cell = p_grid[x][y]
            if "playerWhoPlanted" in cell:
                out += p_grid[x][y]["playerWhoPlanted"][-1]
                out += p_grid[x][y]["plantType"][0]
            else:
                out += ". "
        out += "\n"
    return out
```

```python
from termcolor import colored
# print(colored('hello', 'red'), colored('world', 'green'))

COLOR_MAP = {"1": "blue", "2": "green", "e": "red"}    
def str2cstr(p_str):
    if p_str in COLOR_MAP:
        return colored(p_str,COLOR_MAP[p_str],attrs=['reverse'])
    return p_str

def grid2cstr(p_grid):
    out = ""
    for y in yrange:
        for x in xrange:
            cell = p_grid[x][y]
            if "playerWhoPlanted" in cell:
                out += str2cstr(p_grid[x][y]["playerWhoPlanted"][-1])
                out += str2cstr(p_grid[x][y]["plantType"][0])
            else:
                out += str2cstr(". ")
        out += "\n"
    return out
```
